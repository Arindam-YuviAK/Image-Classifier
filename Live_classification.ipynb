```python
from keras.models import load_model
import cv2
import numpy as np
import tensorflow as tf
```

```python

# font
font = cv2.FONT_HERSHEY_SIMPLEX
  
# org
org = (10, 50)
  
# fontScale
fontScale = 1
   
# Yellow color in BGR
color = (0,255,255)
  
# Line thickness of 2 px
thickness = 2

text = 'Box = '

# Box Orientation
x1,y1,w1,h1 = 0,0,235,75
```

```python
model = load_model('models/BOXclassifier_1.h5')

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Can't open Webcam / Video")

    
while True:
    ret, img = cap.read()
    resize = tf.image.resize(img, (256,256))
    yhat = model.predict(np.expand_dims(resize/255, 0))

    if yhat > 0.5: 
        print(f'Predicted class is No_Box')
        cv2.rectangle(img, (x1, x1), (x1 + w1, y1 + h1), (0,0,0), -1)
        cv2.putText(img, "No_Box", org, font, fontScale, color, thickness, cv2.LINE_AA)
    else:
        print(f'Predicted class is Box')
        cv2.rectangle(img, (x1, x1), (x1 + w1, y1 + h1), (0,0,0), -1)
        cv2.putText(img, "Box", org, font, fontScale, color, thickness, cv2.LINE_AA)
         
    cv2.imshow('BOX Detection', img)
    count = 0
    if cv2.waitKey(2) & 0xFF == ord('x'):
        break

cap.release()
cv2.destroyAllWindows()
```
